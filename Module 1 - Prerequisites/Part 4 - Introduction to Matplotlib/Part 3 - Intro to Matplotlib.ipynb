<h3><div id='1'> What is Matplotlib? </div></h3>
<br /> 
Matplotlib is a low level graph plotting library in python that serves as a visualization utility.

Matplotlib was created by John D. Hunter.

Matplotlib is open source and we can use it freely.

Matplotlib is mostly written in python, a few segments are written in C, Objective-C and Javascript for Platform compatibility.

ref: https://www.w3schools.com/python/matplotlib_intro.asp
<br/>

Click <a href='https://matplotlib.org/stable/tutorials/index.html'>here</a> to read more about.

<hr />

<h3><div id='2'>Why it is used?</div></h3>
<br />
Matplotlib is a cross-platform, data visualization and graphical plotting library for Python and its numerical extension NumPy. As such, it offers a viable open source alternative to MATLAB. Developers can also use matplotlib's APIs (Application Programming Interfaces) to embed plots in GUI applications.

<h3>Pyplot</h3>

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
```

### `plot`

Plot function is the marker that you can draw your charts with.

plot() without any configuration connects points using a line. 

```python
# without x_axis 
points_y = np.array([2, 3, 4, 123])

plt.plot(points_y)
plt.show()
```

plot the chart with the given x_axis:

```python
# plot(x_axis, y_axis)

points_x = np.array([1, 2, 3, 4, 5, 6, 7, 8])
points_y = np.array([1, 8, 8, 1, 1, 8, 8, 1])

plt.plot(points_x, points_y)
plt.show()
```

Plot without connecting points to each other: 

```python
plt.plot(points_y, 'o')
plt.show()
```

Changing the points marker:

```python
plt.plot(points_y, 'x')
plt.show()
```

```python
plt.plot(points_y, marker='*')
plt.show()
```

how to change the design of the plot and make it personalize :

```python
plt.plot(points_y, 'o-.r')
plt.show()
```

<h3>Labels</h3>

You can set labels to the whole plot, y_axis and x_axis.

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.show()
```

### `Grid`

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage")

plt.plot(x, y)
plt.grid()
plt.show()
```

<h3>Subplot</h3>

Put plots beside or top of each other. 

```python
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)

plt.show()
```

<h3>Bar chart</h3>

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x, y, height = 0.1)
plt.show()
```

```python
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, width = 0.6,  color = "#4CAF50")
plt.show()
```

<h3>Pie chart</h3>

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])

mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode)

plt.show() 
```
