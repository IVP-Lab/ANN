<div id='1'><h2>Momentum</h2></div>
<p>Momentum is an extension to the gradient descent optimization algorithm that allows the search to build inertia in a direction in the search space and overcome the oscillations of noisy gradients and coast across flat spots of the search space.</p>

<div id='1'><h2>Learnining Rate</h2></div>
<p>The learning rate is a hyperparameter that controls how much to change the model in response to the estimated error each time the model weights are updated. Choosing the learning rate is challenging as a value too small may result in a long training process that could get stuck, whereas a value too large may result in learning a sub-optimal set of weights too fast or an unstable training process.</p>

```python
# These are the libraries that will be used for this lab.

import torch 
import torch.nn as nn
import numpy as np
import matplotlib.pyplot as plt

torch.manual_seed(5)
```

```python
points = np.arange(-10, 10, 0.1)
squared_points = points ** 2

plt.plot(points, squared_points)
plt.title("y = x^2")
```

```python
square = lambda a : a ** 2

class OneParam(nn.Module):
    def __init__(self, input_size, output_size):
        super(OneParam, self).__init__()
        self.linear = nn.Linear(input_size, output_size, bias=False)
        
    def forward(self, x):
        yhat = self.linear(x)
        return yhat
```

```python
def create_and_train_square(lr, momentum,  n_epochs=10):
  one_param = OneParam(1, 1)
  optimizer = torch.optim.SGD(one_param.parameters(), lr=lr, momentum=momentum)

  one_param.state_dict()['linear.weight'][0] = 9.0

  parameter = []
  loss_list = []

  for n in range(n_epochs):
      optimizer.zero_grad() 
      loss = square(one_param(torch.tensor([[1.0]])))
      loss_list.append(loss)
      parameter.append(one_param.state_dict()['linear.weight'][0].detach().data.item())
      loss.backward()
      optimizer.step()

  plt.plot(parameter, loss_list, 'ro', label='parameter values')
  plt.plot(points, squared_points, label='objective function')
  plt.xlabel('w')
  plt.ylabel('l(w)')
  plt.legend()
```

```python
create_and_train_square(lr=0.01, momentum=0)
```
