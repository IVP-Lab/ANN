<h2><div id='1'>Libraries & Contants</div></h2>

```python
import torch
import torch.nn as nn
```

<h2><div id='2'>One dimension Convolutional layer</div></h2>
When you are designing and implementing a network, the most significant tas is searching and reading docs to realize how it is possible to be implemented it in your way. Please take a glance at this <a href="https://pytorch.org/docs/stable/generated/torch.nn.Conv1d.html">link</a>.

```python
conv1d = nn.Conv1d(16, 33, 3, stride=2)
input = torch.randn(20, 16, 50)
print("input shape is:", input.shape)

output = conv1d(input)
print("output shape is:", output.shape)
```

<h2><div id='3'>Two dimension Convolutional layer</div></h2>
Please take a glance at this <a href="https://pytorch.org/docs/stable/generated/torch.nn.Conv2d.html">link</a>.


```python
# With square kernels and equal stride
conv2d = nn.Conv2d(16, 33, 3, stride=2)

# non-square kernels and unequal stride and with padding
conv2d = nn.Conv2d(16, 33, (3, 5), stride=(2, 1), padding=(4, 2))

input = torch.randn(20, 16, 50, 100)
print("input shape is:", input.shape)

output = conv2d(input)
print("output shape is:", output.shape)
```

<h2><div id='4'>Three dimension Convolutional layer</div></h2>
Please take a glance at this <a href="https://pytorch.org/docs/1.9.1/generated/torch.nn.Conv3d.html">link</a>.


```python
# With square kernels and equal stride
conv3d = nn.Conv3d(16, 33, 3, stride=2)

# non-square kernels and unequal stride and with padding
conv3d = nn.Conv3d(16, 33, (3, 5, 2), stride=(2, 1, 1), padding=(4, 2, 0))

input = torch.randn(20, 16, 10, 50, 100)
print("input shape is:", input.shape)

output = conv3d(input)
print("output shape is:", output.shape)
```

<h2><div id='5'>One dimension Maxpool layer</div></h2>
Please take a glance at this <a href="https://pytorch.org/docs/stable/generated/torch.nn.MaxPool1d.html">link</a>.


```python
# pool of size=3, stride=2
maxPool1d = nn.MaxPool1d(3, stride=2)
input = torch.randn(20, 16, 50)
output = maxPool1d(input)
```

<h2><div id='6'>Two dimension Maxpool layer</div></h2>
Please take a glance at this <a href="https://pytorch.org/docs/stable/generated/torch.nn.MaxPool2d.html">link</a>.



```python
m = nn.MaxPool2d(3, stride=2)
# pool of non-square window
m = nn.MaxPool2d((3, 2), stride=(2, 1))
input = torch.randn(20, 16, 50, 32)
output = m(input)
```

<h2><div id='7'>Two dimension Average Pooling layer</div></h2>
Please take a glance at this <a href="https://pytorch.org/docs/stable/generated/torch.nn.AvgPool2d.html">link</a>.


```python
# pool of square window of size=3, stride=2
m = nn.AvgPool2d(3, stride=2)
# pool of non-square window
m = nn.AvgPool2d((3, 2), stride=(2, 1))
input = torch.randn(20, 16, 50, 32)
output = m(input)
```
